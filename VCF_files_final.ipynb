```python
import pandas as pd
import os

def encode_deletions_from_vcfs(vcf_paths):
    result = {}

    for vcf_path in vcf_paths:
        deletions = []
        tool_name = os.path.splitext(os.path.basename(vcf_path))[0]


        with open(vcf_path, 'r') as f:
            for line in f:
                if line.startswith('#'):
                    continue
                fields = line.strip().split('\t')
                chrom = fields[0]
                pos = int(fields[1])
                ref = fields[3]
                alt = fields[4]
                info = fields[7]

                if 'SVTYPE=DEL' in info:
                    svlen = None
                    end = None

                    for entry in info.split(';'):
                        if entry.startswith('SVLEN='):
                            try:
                                svlen = abs(int(entry.split('=')[1]))
                            except:
                                svlen = None
                        elif entry.startswith('END='):
                            try:
                                end = int(entry.split('=')[1])
                            except:
                                end = None
                    if svlen is None and end is not None:
                        svlen = abs(end - pos)

                    if svlen:
                        deletions.append({
                            'CHROM': chrom,
                            'POS': pos,
                            'SVLEN': svlen
                        })

            result[tool_name] = pd.DataFrame(deletions)



    return result
```

```python
import pysam
import cv2
import numpy as np
from collections import defaultdict

class ImageEncoder:

    def encode_to_image(self, row_id, image, image_start, start, end, sv_len, encoded_row):
        if image_start >= start: # if read starts before start of 3 x sv_len window
            encode_start = image_start-start
            encode_end = image_start-start+sv_len*3
            encoded_image = encoded_row[encode_start:encode_end]
            image[row_id, 0:encoded_image.shape[0], :] = encoded_image

        else:
            encode_start = abs(image_start-start)
            encode_end = sv_len*3 - encode_start
            encoded_image = encoded_row[0:encode_end]
            image[row_id, encode_start:encode_start+encoded_image.shape[0], :] = encoded_image

        return image

    def encode(self, chr_, bp, bam_file, mutation_type="DEL"):
        sv_len = bp[1] - bp[0]

        rows_dict = defaultdict(list)

        #iterate over reads mapped to mutation region
        for read in bam_file.fetch(chr_, bp[0], bp[1], multiple_iterators=True):
            # filterout low quality reads
            if read.mapping_quality < 60:
                continue

            counter = 0

            #initialize empty array (zeros - black color) depends of read lenght
            row = np.zeros((read.reference_end-read.reference_start, 3))

            for ct in read.cigartuples:
                # Red color as a mapped nucleotides
                if ct[0] in [0]:
                    row[counter:counter+ct[1], :] = [1, 0, 0]
                    counter += ct[1]
                # Green color as a deletions
                elif ct[0] in [2]:
                    row[counter:counter+ct[1], :] = [0, 1, 0]
                    counter += ct[1]
                # Blue color as a insterions. We encode insertions only for mutations_type=INS and only in mutation region, not in extended space
                elif ct[0] in [1] and (counter >= bp[0] - read.reference_start - sv_len and counter <= bp[1] - read.reference_start + sv_len) and mutation_type == "INS":
                    row[counter:counter+ct[1], :] = [0, 0, 1]
                    counter += ct[1]

            rows_dict[read.query_name].append([row, read.reference_start,read.reference_end])

        if not rows_dict:
            return None
        #print(len(rows_dict), bp)
        #initialize variant image with zeros
        image = np.zeros(
            (
                len(rows_dict.keys()),
                sv_len*3,
                3
            )
        )

        # image start relative to position on chromosome
        image_start = bp[0] - sv_len

        row_id = 0
        for key, value in rows_dict.items():
            # encode only reads mapped to single region
            if len(value) == 1:
                encoded_row, start, end = value[0]
                image = self.encode_to_image(row_id, image, image_start, start, end, sv_len, encoded_row)
                row_id += 1
            elif len(value) == 2:
                value = sorted(value, key=lambda x: x[1])
                el1 = value[0]
                el2 = value[1]
                image = self.encode_to_image(row_id, image, image_start, el1[1], el1[2], sv_len, el1[0])
                image = self.encode_to_image(row_id, image, image_start, el2[1], el2[2], sv_len, el2[0])
                image[row_id, el1[2]-image_start:el2[1]-image_start, :] = [0, 1, 0]
                row_id += 1
        # if less than two valid reads mapped - return None
        if row_id < 2:
            return None

        resized_image = cv2.resize(image[:row_id, :], (50, 50))
        return np.array(resized_image)
```

```python

VCF_SNIFF  = "/content/drive/MyDrive/Task2/sniffles_chr21.vcf"
VCF_CUTESV = "/content/drive/MyDrive/Task2/cutesv_chr21.vcf"
VCF_DYSGU  = "/content/drive/MyDrive/Task2/result.vcf"
BAM_PATH   = "/content/drive/MyDrive/Task2/HG002_10X_R9_CHR21_sorted.bam"
MAX_IMAGES = 800


import pysam, random, numpy as np, matplotlib.pyplot as plt
from tqdm import tqdm

def get_dels(vcf, tag):
    for r in pysam.VariantFile(vcf):
        if r.info.get("SVTYPE")=="DEL":
            yield dict(caller=tag, chrom=r.chrom, start=r.pos, end=r.stop)

var = [*get_dels(VCF_SNIFF,"Sniffles"),
       *get_dels(VCF_CUTESV,"cuteSV"),
       *get_dels(VCF_DYSGU ,"Dysgu")]
var = random.sample(var, k=min(MAX_IMAGES,len(var)))

enc = ImageEncoder(); bam = pysam.AlignmentFile(BAM_PATH,"rb")
imgs, labs = [], []
for v in tqdm(var,desc="encode"):
    img = enc.encode(v["chrom"],(v["start"],v["end"]),bam,"DEL")
    if img is not None:
        imgs.append(img.astype("float32").ravel())
        labs.append(v["caller"])
bam.close(); X = np.stack(imgs)


from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap.umap_ as umap

emb = { "PCA" : PCA (2).fit_transform(X),
        "t-SNE": TSNE(2, init="pca", learning_rate="auto").fit_transform(X),
        "UMAP": umap.UMAP(2).fit_transform(X) }

clr = {"Sniffles":"tab:red", "cuteSV":"tab:blue", "Dysgu":"tab:green"}

def plot(e, title):
    plt.figure(figsize=(4,4))
    for k,c in clr.items():
        m = [i for i,l in enumerate(labs) if l==k]
        plt.scatter(e[m,0], e[m,1], c=c, s=12, alpha=.75, label=k)
    plt.title(title); plt.legend(fontsize=8); plt.xticks([]); plt.yticks([])
    plt.tight_layout(); plt.show()

for t,e in emb.items(): plot(e, t)
```

```python
%%bash
set -euo pipefail
echo "==> start $(date)"

WORK=/content
GT_DIR=/content/drive/MyDrive/Task2
cd "$WORK"

# 0) compress, index i usuń konfliktowe tagi
for base in sniffles_chr21 cutesv_chr21 result21; do
  raw_vcf="${WORK}/${base}.vcf"
  raw_vcfgz="${WORK}/${base}.vcf.gz"
  clean="${WORK}/${base}_clean.vcf.gz"

  # 0a) bgzip/tabix jeśli trzeba
  if [[ ! -f "$clean" ]]; then
    if [[ -f "$raw_vcf" ]]; then
      echo "--- Compressing $raw_vcf → $clean"
      bgzip -c "$raw_vcf" > "$clean"
    elif [[ -f "$raw_vcfgz" ]]; then
      echo "--- Copying $raw_vcfgz → $clean"
      cp "$raw_vcfgz" "$clean"
    else
      echo "ERROR: missing ${base}.vcf(.gz)" >&2
      exit 1
    fi
    tabix -p vcf "$clean"
  else
    echo "+++ $clean exists, skipping compression"
  fi

  # 0b) usuń konfliktowe INFO-tag’i
  echo "--- Stripping INFO/STRAND,INFO/GQ from $clean"
  bcftools annotate -x INFO/STRAND,INFO/GQ -Oz -o tmp.vcf.gz "$clean"
  mv tmp.vcf.gz "$clean"
  tabix -f -p vcf "$clean"
done

# 1) ground-truth też musi być bgzip+tabix, ale nie stripujemy
GT_RAW="${GT_DIR}/HG002_GROUND_TRUTH_CHR21_DEL.vcf"
GT_CLEAN="${WORK}/HG002_GROUND_TRUTH_CHR21_DEL_clean.vcf.gz"
if [[ ! -f "$GT_CLEAN" ]]; then
  echo "--- Compressing GT $GT_RAW → $GT_CLEAN"
  bgzip -c "$GT_RAW" > "$GT_CLEAN"
  tabix -p vcf "$GT_CLEAN"
else
  echo "+++ $GT_CLEAN exists, skipping"
fi

# FASTA i indeks
REF="${GT_DIR}/chr21.fa"
if [[ ! -f "${REF}.fai" ]]; then
  echo "Indexing FASTA…"
  samtools faidx "$REF"
fi

# 2) MERGE callerów
echo "== MERGE =="
bcftools merge -m none --force-samples -O z \
  -o merged_calls.vcf.gz \
  "${WORK}/sniffles_chr21_clean.vcf.gz" \
  "${WORK}/cutesv_chr21_clean.vcf.gz" \
  "${WORK}/result21_clean.vcf.gz"
tabix -f -p vcf merged_calls.vcf.gz

# 3) COLLAPSE
echo "== COLLAPSE =="
truvari collapse \
  -i merged_calls.vcf.gz \
  -o collapsed_calls.vcf.gz
tabix -f -p vcf collapsed_calls.vcf.gz

# 4) BENCH
echo "== BENCH =="
truvari bench \
  -b "${WORK}/HG002_GROUND_TRUTH_CHR21_DEL_clean.vcf.gz" \
  -c collapsed_calls.vcf.gz \
  -f "$REF" \
  -o bench_chr22 \
  --passonly -r 500 -p 0.0

echo "✓ Finished – results in bench_chr21/"
```

```python
import pysam
import numpy as np
import pandas as pd

# Open BAM files
bam_train = pysam.AlignmentFile("/content/HG002_30X_R9_CHR21_sorted.bam", "rb")
bam_test  = pysam.AlignmentFile("/content/HG002_30X_R9_CHR22_sorted.bam", "rb")

# Initialize image encoder
encoder = ImageEncoder()

def load_and_encode(csv_path, label, bam, encoder, mutation_type="DEL", window=200):
    """
    Load variant CSV, fetch reads from BAM, encode to image, return lists of images and labels.
    """
    df = pd.read_csv(csv_path)
    X, y = [], []
    half = window // 2

    for _, row in df.iterrows():
        chrom = str(row['chrom'])
        start = int(row['pos'])

        # Determine end coordinate
        if pd.isna(row.get('end')):
            if pd.notna(row.get('svlen')):
                svlen = int(row['svlen'])
                end = start + abs(svlen)
            else:
                end = start
        else:
            end = int(row['end'])

        # Build flanked interval ensuring start <= end
        left = max(min(start, end) - half, 0)
        right = max(max(start, end) + half, left)

        # Encode region to image
        try:
            img = encoder.encode(chrom, (left, right), bam, mutation_type)
        except Exception as e:
            print(f"Skipping region {chrom}:{left}-{right} due to error: {e}")
            continue

        if img is not None:
            X.append((img * 255).astype(np.uint8))
            y.append(label)

    return X, y

# Load TP/FN examples for chr21 and chr22
X_train_tp, y_train_tp = load_and_encode("tp_chr21.csv", 1, bam_train, encoder)
X_train_fn, y_train_fn = load_and_encode("fp_chr21.csv", 0, bam_train, encoder)
X_test_tp,  y_test_tp  = load_and_encode("tp_chr22.csv", 1, bam_test,  encoder)
X_test_fn,  y_test_fn  = load_and_encode("fp_chr22.csv", 0, bam_test,  encoder)

# Determine image shape from first non-empty
img_shape = None
for chunk in (X_train_tp, X_train_fn, X_test_tp, X_test_fn):
    if chunk:
        img_shape = chunk[0].shape
        break
if img_shape is None:
    raise ValueError("No images loaded; check CSV paths and BAM files.")

# Helper to stack safely

def safe_stack(lst):
    """Stack list of images to array shape (n, H, W, C) or return empty of proper shape."""
    if not lst:
        return np.empty((0, *img_shape), dtype=np.uint8)
    return np.stack(lst, axis=0)

# Build final arrays
X_train = safe_stack(X_train_tp + X_train_fn)
y_train = np.array(y_train_tp + y_train_fn, dtype=int)
X_test  = safe_stack(X_test_tp  + X_test_fn)
y_test  = np.array(y_test_tp  + y_test_fn, dtype=int)

# Save
np.save("X_train.npy", X_train)
pd.DataFrame(y_train, columns=["label"]).to_csv("y_train.csv", index=False)
np.save("X_test.npy",  X_test)
pd.DataFrame(y_test, columns=["label"]).to_csv("y_test.csv",  index=False)

# Visualize embeddings
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap

# Flatten for projections
X_all = np.concatenate([X_train, X_test], axis=0)
y_all = np.concatenate([y_train, y_test], axis=0)
flat = X_all.reshape((X_all.shape[0], -1))

# PCA
pca = PCA(n_components=2)
proj_pca = pca.fit_transform(flat)
plt.figure(figsize=(8,6))
plt.scatter(proj_pca[:,0], proj_pca[:,1], c=y_all, cmap='coolwarm', alpha=0.7)
plt.title("PCA: TP vs FN")
plt.colorbar(label="Label")
plt.show()

# t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
proj_tsne = tsne.fit_transform(flat)
plt.figure(figsize=(8,6))
plt.scatter(proj_tsne[:,0], proj_tsne[:,1], c=y_all, cmap='coolwarm', alpha=0.7)
plt.title("t-SNE: TP vs FN")
plt.colorbar(label="Label")
plt.show()

# UMAP
umap_model = umap.UMAP(n_components=2, random_state=42)
proj_umap = umap_model.fit_transform(flat)
plt.figure(figsize=(8,6))
plt.scatter(proj_umap[:,0], proj_umap[:,1], c=y_all, cmap='coolwarm', alpha=0.7)
plt.title("UMAP: TP vs FN")
plt.colorbar(label="Label")
plt.show()
```
