```python
import pandas as pd
import os

def encode_deletions_from_vcfs(vcf_paths):
    result = {}

    for vcf_path in vcf_paths:
        deletions = []
        tool_name = os.path.splitext(os.path.basename(vcf_path))[0]


        with open(vcf_path, 'r') as f:
            for line in f:
                if line.startswith('#'):
                    continue
                fields = line.strip().split('\t')
                chrom = fields[0]
                pos = int(fields[1])
                ref = fields[3]
                alt = fields[4]
                info = fields[7]

                if 'SVTYPE=DEL' in info:
                    svlen = None
                    end = None

                    for entry in info.split(';'):
                        if entry.startswith('SVLEN='):
                            try:
                                svlen = abs(int(entry.split('=')[1]))
                            except:
                                svlen = None
                        elif entry.startswith('END='):
                            try:
                                end = int(entry.split('=')[1])
                            except:
                                end = None
                    if svlen is None and end is not None:
                        svlen = abs(end - pos)

                    if svlen:
                        deletions.append({
                            'CHROM': chrom,
                            'POS': pos,
                            'SVLEN': svlen
                        })

            result[tool_name] = pd.DataFrame(deletions)



    return result
```

```python
import pysam
import cv2
import numpy as np
from collections import defaultdict

class ImageEncoder:

    def encode_to_image(self, row_id, image, image_start, start, end, sv_len, encoded_row):
        if image_start >= start: # if read starts before start of 3 x sv_len window
            encode_start = image_start-start
            encode_end = image_start-start+sv_len*3
            encoded_image = encoded_row[encode_start:encode_end]
            image[row_id, 0:encoded_image.shape[0], :] = encoded_image

        else:
            encode_start = abs(image_start-start)
            encode_end = sv_len*3 - encode_start
            encoded_image = encoded_row[0:encode_end]
            image[row_id, encode_start:encode_start+encoded_image.shape[0], :] = encoded_image

        return image

    def encode(self, chr_, bp, bam_file, mutation_type="DEL"):
        sv_len = bp[1] - bp[0]

        rows_dict = defaultdict(list)

        #iterate over reads mapped to mutation region
        for read in bam_file.fetch(chr_, bp[0], bp[1], multiple_iterators=True):
            # filterout low quality reads
            if read.mapping_quality < 60:
                continue

            counter = 0

            #initialize empty array (zeros - black color) depends of read lenght
            row = np.zeros((read.reference_end-read.reference_start, 3))

            for ct in read.cigartuples:
                # Red color as a mapped nucleotides
                if ct[0] in [0]:
                    row[counter:counter+ct[1], :] = [1, 0, 0]
                    counter += ct[1]
                # Green color as a deletions
                elif ct[0] in [2]:
                    row[counter:counter+ct[1], :] = [0, 1, 0]
                    counter += ct[1]
                # Blue color as a insterions. We encode insertions only for mutations_type=INS and only in mutation region, not in extended space
                elif ct[0] in [1] and (counter >= bp[0] - read.reference_start - sv_len and counter <= bp[1] - read.reference_start + sv_len) and mutation_type == "INS":
                    row[counter:counter+ct[1], :] = [0, 0, 1]
                    counter += ct[1]

            rows_dict[read.query_name].append([row, read.reference_start,read.reference_end])

        if not rows_dict:
            return None
        #print(len(rows_dict), bp)
        #initialize variant image with zeros
        image = np.zeros(
            (
                len(rows_dict.keys()),
                sv_len*3,
                3
            )
        )

        # image start relative to position on chromosome
        image_start = bp[0] - sv_len

        row_id = 0
        for key, value in rows_dict.items():
            # encode only reads mapped to single region
            if len(value) == 1:
                encoded_row, start, end = value[0]
                image = self.encode_to_image(row_id, image, image_start, start, end, sv_len, encoded_row)
                row_id += 1
            elif len(value) == 2:
                value = sorted(value, key=lambda x: x[1])
                el1 = value[0]
                el2 = value[1]
                image = self.encode_to_image(row_id, image, image_start, el1[1], el1[2], sv_len, el1[0])
                image = self.encode_to_image(row_id, image, image_start, el2[1], el2[2], sv_len, el2[0])
                image[row_id, el1[2]-image_start:el2[1]-image_start, :] = [0, 1, 0]
                row_id += 1
        # if less than two valid reads mapped - return None
        if row_id < 2:
            return None

        resized_image = cv2.resize(image[:row_id, :], (50, 50))
        return np.array(resized_image)
```

```python
%%bash
set -euo pipefail
echo "==> start $(date)"

WORK=/content
GT_DIR=/content/drive/MyDrive/Task2
cd "$WORK"

# 0) compress, index i usuń konfliktowe tagi
for base in sniffles_chr21 cutesv_chr21 result21; do
  raw_vcf="${WORK}/${base}.vcf"
  raw_vcfgz="${WORK}/${base}.vcf.gz"
  clean="${WORK}/${base}_clean.vcf.gz"

  # 0a) bgzip/tabix jeśli trzeba
  if [[ ! -f "$clean" ]]; then
    if [[ -f "$raw_vcf" ]]; then
      echo "--- Compressing $raw_vcf → $clean"
      bgzip -c "$raw_vcf" > "$clean"
    elif [[ -f "$raw_vcfgz" ]]; then
      echo "--- Copying $raw_vcfgz → $clean"
      cp "$raw_vcfgz" "$clean"
    else
      echo "ERROR: missing ${base}.vcf(.gz)" >&2
      exit 1
    fi
    tabix -p vcf "$clean"
  else
    echo "+++ $clean exists, skipping compression"
  fi

  # 0b) usuń konfliktowe INFO-tag’i
  echo "--- Stripping INFO/STRAND,INFO/GQ from $clean"
  bcftools annotate -x INFO/STRAND,INFO/GQ -Oz -o tmp.vcf.gz "$clean"
  mv tmp.vcf.gz "$clean"
  tabix -f -p vcf "$clean"
done

# 1) ground-truth też musi być bgzip+tabix, ale nie stripujemy
GT_RAW="${GT_DIR}/HG002_GROUND_TRUTH_CHR21_DEL.vcf"
GT_CLEAN="${WORK}/HG002_GROUND_TRUTH_CHR21_DEL_clean.vcf.gz"
if [[ ! -f "$GT_CLEAN" ]]; then
  echo "--- Compressing GT $GT_RAW → $GT_CLEAN"
  bgzip -c "$GT_RAW" > "$GT_CLEAN"
  tabix -p vcf "$GT_CLEAN"
else
  echo "+++ $GT_CLEAN exists, skipping"
fi

# FASTA i indeks
REF="${GT_DIR}/chr21.fa"
if [[ ! -f "${REF}.fai" ]]; then
  echo "Indexing FASTA…"
  samtools faidx "$REF"
fi

# 2) MERGE callerów
echo "== MERGE =="
bcftools merge -m none --force-samples -O z \
  -o merged_calls.vcf.gz \
  "${WORK}/sniffles_chr21_clean.vcf.gz" \
  "${WORK}/cutesv_chr21_clean.vcf.gz" \
  "${WORK}/result21_clean.vcf.gz"
tabix -f -p vcf merged_calls.vcf.gz

# 3) COLLAPSE
echo "== COLLAPSE =="
truvari collapse \
  -i merged_calls.vcf.gz \
  -o collapsed_calls.vcf.gz
tabix -f -p vcf collapsed_calls.vcf.gz

# 4) BENCH
echo "== BENCH =="
truvari bench \
  -b "${WORK}/HG002_GROUND_TRUTH_CHR21_DEL_clean.vcf.gz" \
  -c collapsed_calls.vcf.gz \
  -f "$REF" \
  -o bench_chr22 \
  --passonly -r 500 -p 0.0

echo "✓ Finished – results in bench_chr21/"
```

```python
%%bash
set -euo pipefail
echo "==> start $(date)"

WORK=/content
DATA=/content/drive/MyDrive/Task2

# 1) sort & index BAMs
samtools sort -o ${WORK}/HG002_30X_R9_CHR21_sorted.bam ${DATA}/HG002_30X_R9_CHR21.bam
samtools sort -o ${WORK}/HG002_30X_R9_CHR22_sorted.bam ${DATA}/HG002_30X_R9_CHR22.bam
samtools index ${WORK}/HG002_30X_R9_CHR21_sorted.bam
samtools index ${WORK}/HG002_30X_R9_CHR22_sorted.bam
# 2) SV calling

# Dysgu chr21
rm -rf /tmp/dysgu_out21 && mkdir -p /tmp/dysgu_out21
dysgu run -x ${DATA}/chr21.fa /tmp/dysgu_out21 \
    ${WORK}/HG002_30X_R9_CHR21_sorted.bam > ${WORK}/result21.vcf

# Dysgu chr22
rm -rf /tmp/dysgu_out22 && mkdir -p /tmp/dysgu_out22
dysgu run -x ${DATA}/chr22.fa /tmp/dysgu_out22 \
    ${WORK}/HG002_30X_R9_CHR22_sorted.bam > ${WORK}/result22.vcf

# Sniffles chr21/22 (with overwrite)
sniffles -i ${WORK}/HG002_30X_R9_CHR21_sorted.bam \
         -v ${WORK}/sniffles_chr21.vcf \
         --reference ${DATA}/chr21.fa \
         --allow-overwrite

sniffles -i ${WORK}/HG002_30X_R9_CHR22_sorted.bam \
         -v ${WORK}/sniffles_chr22.vcf \
         --reference ${DATA}/chr22.fa \
         --allow-overwrite

# cuteSV chr21
rm -rf /tmp/cute21 && mkdir -p /tmp/cute21
cuteSV ${WORK}/HG002_30X_R9_CHR21_sorted.bam ${DATA}/chr21.fa \
       ${WORK}/cutesv_chr21.vcf /tmp/cute21

# cuteSV chr22
rm -rf /tmp/cute22 && mkdir -p /tmp/cute22
cuteSV ${WORK}/HG002_30X_R9_CHR22_sorted.bam ${DATA}/chr22.fa \
       ${WORK}/cutesv_chr22.vcf /tmp/cute22

# 3) compress & index + strip conflicts
clean_vcf(){
  raw_vcf="$1"
  base=$(basename "$raw_vcf" .vcf)
  clean="${WORK}/${base}_clean.vcf.gz"

  # 1) bgzip + index
  bgzip -c "$raw_vcf" > "$clean"
  tabix -f -p vcf "$clean"

  # 2) sanitize (normalize columns)
  bcftools view -Ou "$clean" \
    | bcftools view -Oz -o tmp.sanitize.vcf.gz
  tabix -f -p vcf tmp.sanitize.vcf.gz

  # 3) strip conflicting INFO tags
  bcftools annotate -x INFO/STRAND,INFO/GQ -Oz \
    -o tmp.clean.vcf.gz tmp.sanitize.vcf.gz
  tabix -f -p vcf tmp.clean.vcf.gz

  # 4) overwrite and re-index
  mv tmp.clean.vcf.gz "$clean"
  tabix -f -p vcf "$clean"
  rm -f tmp.sanitize.vcf.gz*
}

# then loop:
for v in sniffles_chr21 cutesv_chr21 result21 \
         sniffles_chr22 cutesv_chr22 result22; do
  clean_vcf "${WORK}/${v}.vcf"
done


# 4) Ground-truth
for chr in 21 22; do
  bgzip -c "${DATA}/HG002_GROUND_TRUTH_CHR${chr}_DEL.vcf" \
    > "${WORK}/HG002_GROUND_TRUTH_CHR${chr}_DEL_clean.vcf.gz"
  tabix -f -p vcf "${WORK}/HG002_GROUND_TRUTH_CHR${chr}_DEL_clean.vcf.gz"
done

# 5) Index FASTA if missing
for chr in 21 22; do
  REF=${DATA}/chr${chr}.fa
  [[ -f "${REF}.fai" ]] || samtools faidx "$REF"
done

# 6) Merge → collapse → bench each chr
for chr in 21 22; do
  echo "=== CHR${chr} ==="
  bcftools merge -m none --force-samples -Oz \
    -o merged_chr${chr}.vcf.gz \
      ${WORK}/sniffles_chr${chr}_clean.vcf.gz \
      ${WORK}/cutesv_chr${chr}_clean.vcf.gz \
      ${WORK}/result${chr}_clean.vcf.gz
  tabix -f -p vcf merged_chr${chr}.vcf.gz

  truvari collapse -i merged_chr${chr}.vcf.gz \
    -o collapsed_chr${chr}.vcf.gz
  tabix -f -p vcf collapsed_chr${chr}.vcf.gz

  truvari bench \
    -b ${WORK}/HG002_GROUND_TRUTH_CHR${chr}_DEL_clean.vcf.gz \
    -c collapsed_chr${chr}.vcf.gz \
    -f ${DATA}/chr${chr}.fa \
    -o bench_chr${chr} \
    --passonly -r 500 -p 0.0
done

echo "✓ All done"
```

```python
# Cell 3: build X_train, y_train (chr21) and X_test, y_test (chr22)
import os
import pandas as pd
from cyvcf2 import VCF
import pysam, numpy as np

encoder = ImageEncoder()
window = 200
half   = window // 2

def vcf_to_df(path):
    return pd.DataFrame([{
        "chrom": r.CHROM,
        "pos":   r.POS,
        "end":   r.INFO.get("END"),
        "svlen": r.INFO.get("SVLEN")
    } for r in VCF(path)])

def load_images_from_df(df, bam, label):
    X, y = [], []
    for _, r in df.iterrows():
        start = int(r["pos"])
        # explicit NaN checks
        if pd.notna(r["end"]):
            end = int(r["end"])
        elif pd.notna(r["svlen"]):
            end = start + abs(int(r["svlen"]))
        else:
            end = start

        left  = max(min(start, end) - half, 0)
        right = max(max(start, end) + half, left)
        img   = encoder.encode(str(r["chrom"]), (left, right), bam, "DEL")
        if img is None:
            continue
        X.append((img * 255).astype(np.uint8))
        y.append(label)
    return X, y

# prepare containers
X_train, y_train = [], []
X_test,  y_test  = [], []

# CHR21 → train
bam21     = pysam.AlignmentFile("/content/HG002_30X_R9_CHR21_sorted.bam","rb")
df_tp21   = vcf_to_df("bench_chr21/tp-comp.vcf.gz")
df_fn21   = vcf_to_df("bench_chr21/fp.vcf.gz")
x, y_temp = load_images_from_df(df_tp21, bam21, 1)
X_train.extend(x); y_train.extend(y_temp)
x, y_temp = load_images_from_df(df_fn21, bam21, 0)
X_train.extend(x); y_train.extend(y_temp)

# CHR22 → test
bam22     = pysam.AlignmentFile("/content/HG002_30X_R9_CHR22_sorted.bam","rb")
df_tp22   = vcf_to_df("bench_chr22/tp-comp.vcf.gz")
df_fn22   = vcf_to_df("bench_chr22/fp.vcf.gz")
x, y_temp = load_images_from_df(df_tp22, bam22, 1)
X_test.extend(x);  y_test.extend(y_temp)
x, y_temp = load_images_from_df(df_fn22, bam22, 0)
X_test.extend(x);  y_test.extend(y_temp)

# stack into arrays
X_train = np.stack(X_train, axis=0)
y_train = np.array(y_train, dtype=int)
X_test  = np.stack(X_test,  axis=0)
y_test  = np.array(y_test,  dtype=int)

# optionally save for your neural net
np.save("X_train.npy", X_train)
np.save("y_train.npy", y_train)
np.save("X_test.npy",  X_test)
np.save("y_test.npy",  y_test)

print(f"Shapes → X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"         X_test : {X_test.shape},  y_test : {y_test.shape}")
```
